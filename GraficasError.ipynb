```python
import matplotlib.pyplot as plt
import sys
```

```python
def datosGrafica(archivo,tamaño):
  x, y = [], []
  with open(archivo, "r") as file1:
      datos11 = file1.readlines()
      for i in range(len(datos11)):
        # lectura = file1.readline().split()
        lectura = datos11[i].split()
        y.append(float(lectura[0]))
        x.append(float(lectura[1]))

        lim = 100.0
        if x[i] > lim - tamaño/2 and x[i] < lim + tamaño/2:
          valor = y[i]    
  return x, y, valor


def Graficas(archivos, nombre, titulo):
  tamaños = [3.0, 5.0, 7.0, 10.0]
  for i in range(4):
    x, y, valor = datosGrafica(archivos[i], tamaños[i])
    plt.plot(x, y)
    print(valor)

  plt.xlabel('Distancia [ mm ]')
  plt.ylabel('Temperatura [º C]')
  plt.grid(True)
  plt.legend(['Element size = 03 mm', 'Element size = 05 mm','Element size = 07 mm','Element size = 10 mm'])
  plt.title(titulo)

  ruta = '/content/sample_data/' + nombre
  # plt.savefig(ruta, dpi=600)
  # plt.show()
```

**Graficas pedidas TRIANG-ESTRUCTURADA**

```python
# --- CVFEM ---
archivos = ['CVFEM-trian_03_struc.txt','CVFEM-trian_05_struc.txt','CVFEM-trian_07_struc.txt','CVFEM-trian_10_struc.txt']
nombre = 'CVFEM_trian_structural.png'
titulo = 'CVFEM - Elementos Triangulares Estructurados'

Graficas(archivos, nombre, titulo)
```

```python
# --- FEM ---
archivos = ['FEM-trian_03_struc.txt','FEM-trian_05_struc.txt','FEM-trian_07_struc.txt','FEM-trian_10_struc.txt']
nombre = 'FEM_trian_structural.png'
titulo = 'FEM - Elementos Triangulares Estructurados'

Graficas(archivos, nombre, titulo)
```
